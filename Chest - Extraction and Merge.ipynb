```python
df = pd.read_csv('Data\\Raw Data\\S6\\Pectoralis – chest\\accel.csv')
df
```

```python
df2 = pd.read_csv('Data\\Raw Data\\S6\\Pectoralis – chest\\elec.csv')
df2
```

# Merging data

```python
elec_ts = df2['Timestamp (microseconds)'].to_numpy()
elec_ts
```

```python
accel_ts = df['Timestamp (microseconds)'].to_numpy()
accel_ts
```

```python
for i in range(len(accel_ts)):
    df.at[i, 'Sample (V)'] = df2['Sample (V)'].iloc[np.argmin(abs(accel_ts[i] - elec_ts))]
```

```python
df
```

# Data filtering and adding labels

```python
print(df['Timestamp (microseconds)'][0])
pd.Timestamp(df['Timestamp (microseconds)'][0], unit='us')
```

```python
# These values are obtained from the Meta Data and converted to UTC from Est 

start = pd.Timestamp(year = 2020, month = 10, day = 31, hour = 18, minute = 35, second = 10)
start_ts = int(start.timestamp() * 1e6)

end = pd.Timestamp(year = 2020, month = 10, day = 31, hour = 18, minute = 36, second = 39)
end_ts = int(end.timestamp() * 1e6)

stairs = df.loc[(df['Timestamp (microseconds)'] >= start_ts) & (df['Timestamp (microseconds)'] <= end_ts)]

stairs['Activity_Label'] = 'stairs'
stairs
```

```python
start = pd.Timestamp(year = 2020, month = 10, day = 31, hour = 18, minute = 44, second = 30)
start_ts = int(start.timestamp() * 1e6)

end = pd.Timestamp(year = 2020, month = 10, day = 31, hour = 18, minute = 45, second = 56)
end_ts = int(end.timestamp() * 1e6)

jog = df.loc[(df['Timestamp (microseconds)'] >= start_ts) & (df['Timestamp (microseconds)'] <= end_ts)]

jog['Activity_Label'] = 'jog'
jog
```

```python
start = pd.Timestamp(year = 2020, month = 10, day = 31, hour = 18, minute = 50, second = 45)
start_ts = int(start.timestamp() * 1e6)

end = pd.Timestamp(year = 2020, month = 10, day = 31, hour = 18, minute = 50, second = 55)
end_ts = int(end.timestamp() * 1e6)

skip = df.loc[(df['Timestamp (microseconds)'] >= start_ts) & (df['Timestamp (microseconds)'] <= end_ts)]
skip['Activity_Label'] = 'skipping'
skip
```

```python
start = pd.Timestamp(year = 2020, month = 10, day = 31, hour = 18, minute = 57, second = 30)
start_ts = int(start.timestamp() * 1e6)

end = pd.Timestamp(year = 2020, month = 10, day = 31, hour = 18, minute = 58, second = 11)
end_ts = int(end.timestamp() * 1e6)

kb = df.loc[(df['Timestamp (microseconds)'] >= start_ts) & (df['Timestamp (microseconds)'] <= end_ts)]
kb['Activity_Label'] = 'kettlebell'
kb
```

```python
start = pd.Timestamp(year = 2020, month = 10, day = 31, hour = 19, minute = 1, second = 0)
start_ts = int(start.timestamp() * 1e6)

end = pd.Timestamp(year = 2020, month = 10, day = 31, hour = 19, minute = 2, second = 11)
end_ts = int(end.timestamp() * 1e6)

bb = df.loc[(df['Timestamp (microseconds)'] >= start_ts) & (df['Timestamp (microseconds)'] <= end_ts)]
bb['Activity_Label'] = 'basket ball'
bb
```

```python
merged_df = pd.concat([stairs, jog, skip, kb, bb], ignore_index=True)
```
