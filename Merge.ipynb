```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
hand = pd.read_csv('Data\\Intermediate Data\\hand_concatenated.csv')
leg = pd.read_csv('Data\\Intermediate Data\\leg_concatenated.csv')
chest = pd.read_csv('Data\\Intermediate Data\\chest_concatenated.csv')
```

# Merging leg and hand sensors

```python
leg_hand = leg.copy()
```

```python
leg_hand
```

```python
hand_ts = hand['Timestamp (microseconds)'].to_numpy()
hand_ts
```

```python
leg_ts = leg['Timestamp (microseconds)'].to_numpy()
leg_ts
```

```python
for i in range(len(leg_ts)):
    j = np.argmin(abs(leg_ts[i] - hand_ts))
    leg_hand.at[i, 'Accel X (g)_hand'] = hand['Accel X (g)'].iloc[j]
    leg_hand.at[i, 'Accel Y (g)_hand'] = hand['Accel Y (g)'].iloc[j]
    leg_hand.at[i, 'Accel Z (g)_hand'] = hand['Accel Z (g)'].iloc[j]
    leg_hand.at[i, 'Gyro X (°/s)_hand'] = hand['Gyro X (°/s)'].iloc[j]
    leg_hand.at[i, 'Gyro Y (°/s)_hand'] = hand['Gyro Y (°/s)'].iloc[j]
    leg_hand.at[i, 'Gyro Z (°/s)_hand'] = hand['Gyro Z (°/s)'].iloc[j]
```

```python
leg_hand
```

```python
leg_hand.rename(columns={"Accel X (g)": "Accel X (g)_leg", "Accel Y (g)": "Accel Y (g)_leg", "Accel Z (g)": "Accel Z (g)_leg",
                        "Gyro X (°/s)": "Gyro X (°/s)_leg", "Gyro Y (°/s)": "Gyro Y (°/s)_leg", "Gyro Z (°/s)": "Gyro Z (°/s)_leg",
                        "Activity_Label_hand": "Activity_Label"}, inplace = True)
```

```python
leg_hand
```

# Up-scaling chest data and performing many to one mapping

```python
merged_up = leg_hand.copy()
```

```python
merged_up
```

```python
merged_up_ts = leg_hand['Timestamp (microseconds)'].to_numpy()
merged_up_ts
```

```python
chest_ts = chest['Timestamp (microseconds)'].to_numpy()
chest_ts
```

```python
for i in range(len(merged_up_ts)):
    j = np.argmin(abs(merged_up_ts[i] - chest_ts))
    merged_up.at[i, 'Accel X (g)_chest'] = chest['Accel X (g)'].iloc[j]
    merged_up.at[i, 'Accel Y (g)_chest'] = chest['Accel Y (g)'].iloc[j]
    merged_up.at[i, 'Accel Z (g)_chest'] = chest['Accel Z (g)'].iloc[j]
    merged_up.at[i, 'Sample (V)'] = chest['Sample (V)'].iloc[j]
```

```python
merged_up
```

```python
merged_up = merged_up[['Subject ID','Timestamp (microseconds)', 'Accel X (g)_leg', 'Accel Y (g)_leg', 'Accel Z (g)_leg', 'Gyro X (°/s)_leg',
                     'Gyro Y (°/s)_leg', 'Gyro Z (°/s)_leg', 'Accel X (g)_hand', 'Accel Y (g)_hand', 'Accel Z (g)_hand', 
                     'Gyro X (°/s)_hand', 'Gyro Y (°/s)_hand', 'Gyro Z (°/s)_hand', 'Accel X (g)_chest', 'Accel Y (g)_chest', 
                     'Accel Z (g)_chest', 'Sample (V)','Activity_Label']]
```

```python
merged_up
```

```python
merged_up.to_csv('Data\\filtered_data.csv.csv', sep=',', index = False)
```

# Down-scaling leg-hand data and performing one to one mapping

```python
merged_down = chest.copy()
```

```python
merged_down
```

```python
merged_down_ts = merged_down['Timestamp (microseconds)'].to_numpy()
merged_down_ts
```

```python
leg_hand_ts = leg_hand['Timestamp (microseconds)'].to_numpy()
leg_hand_ts
```

```python
for i in range(len(merged_down)):
    j = np.argmin(abs(merged_down_ts[i] - leg_hand_ts))
    merged_down.at[i, 'Accel X (g)_leg'] = leg_hand['Accel X (g)_leg'].iloc[j]
    merged_down.at[i, 'Accel Y (g)_leg'] = leg_hand['Accel Y (g)_leg'].iloc[j]
    merged_down.at[i, 'Accel Z (g)_leg'] = leg_hand['Accel Z (g)_leg'].iloc[j]
    merged_down.at[i, 'Gyro X (°/s)_leg'] = leg_hand['Gyro X (°/s)_leg'].iloc[j]
    merged_down.at[i, 'Gyro Y (°/s)_leg'] = leg_hand['Gyro Y (°/s)_leg'].iloc[j]
    merged_down.at[i, 'Gyro Z (°/s)_leg'] = leg_hand['Gyro Z (°/s)_leg'].iloc[j]
    merged_down.at[i, 'Accel X (g)_hand'] = leg_hand['Accel X (g)_hand'].iloc[j]
    merged_down.at[i, 'Accel Y (g)_hand'] = leg_hand['Accel Y (g)_hand'].iloc[j]
    merged_down.at[i, 'Accel Z (g)_hand'] = leg_hand['Accel Z (g)_hand'].iloc[j]
    merged_down.at[i, 'Gyro X (°/s)_hand'] = leg_hand['Gyro X (°/s)_hand'].iloc[j]
    merged_down.at[i, 'Gyro Y (°/s)_hand'] = leg_hand['Gyro Y (°/s)_hand'].iloc[j]
    merged_down.at[i, 'Gyro Z (°/s)_hand'] = leg_hand['Gyro Z (°/s)_hand'].iloc[j]
```

```python
merged_down
```

```python
merged_down.rename(columns={"Accel X (g)": "Accel X (g)_chest", "Accel Y (g)": "Accel Y (g)_chest", 
                            "Accel Z (g)": "Accel Z (g)_chest"}, inplace = True)
```

```python
merged_down = merged_down[['Subject ID','Timestamp (microseconds)', 'Accel X (g)_leg', 'Accel Y (g)_leg', 'Accel Z (g)_leg', 'Gyro X (°/s)_leg',
                     'Gyro Y (°/s)_leg', 'Gyro Z (°/s)_leg', 'Accel X (g)_hand', 'Accel Y (g)_hand', 'Accel Z (g)_hand', 
                     'Gyro X (°/s)_hand', 'Gyro Y (°/s)_hand', 'Gyro Z (°/s)_hand', 'Accel X (g)_chest', 'Accel Y (g)_chest', 
                     'Accel Z (g)_chest', 'Sample (V)','Activity_Label']]
```
